## Introducción
En este notebook, se analizará el desempeño de estudiantes de educación secundaria en dos asignaturas: Matemáticas y Lengua Portuguesa. El dataset, obtenido de [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Student+Performance), contiene datos sobre logros académicos, así como características demográficas, sociales y de entorno familiar de los estudiantes. Este análisis tiene como objetivo identificar patrones y factores que pueden influir en el desempeño académico.

En particular, nos centraremos en entender la relación entre las características de los estudiantes y sus notas en cada período, buscando aquellas variables que más impactan en el rendimiento académico.

## Objetivos
- Realizar un análisis exploratorio de las variables categóricas y numéricas.
- Estudiar la correlación entre las diferentes características y las notas de los estudiantes.
- Visualizar las distribuciones y relaciones clave mediante gráficos.

## <u>Descripción y Clasificación de Variables</u>

### **Cualitativas o Categóricas**

##### **Nominales:**
1. **school** - Escuela del estudiante - ('GP' Gabriel Pereira o 'MS' Mousinho da Silveira)
2. **sex** - Sexo del estudiante - (F o M)
3. **address** - Tipo de dirección del hogar - ('U' - urbano o 'R' - rural)
4. **famsize** - Tamaño de la familia - ('LE3' - menor o igual a 3 o 'GT3' - mayor a 3)
5. **Pstatus** - Estado de convivencia de los padres - ('T' - juntos o 'A' - separados)
6. **Mjob** - Ocupación de la madre - ('teacher', 'health', 'services', 'at_home' u 'other')
7. **Fjob** - Ocupación del padre - ('teacher', 'health', 'services', 'at_home' u 'other')
8. **reason** - Razón para escoger esta escuela - ('home', 'reputation', 'course' u 'other')
9. **guardian** - Tutor del estudiante - ('mother', 'father' u 'other')
10. **schoolsup** - Apoyo educativo extra - (1: <15 min, 2: 15 a 30 min, 3: 30 min a 1 h, 4: >1 h)
11. **famsup** - Apoyo educativo de la familia - ('yes' o 'no')
12. **paid** - Clases extras dentro del curso - ('yes' o 'no')
13. **activities** - Actividades extracurriculares- ('yes' o 'no')
14. **nursery** - Asistió a guardería - ('yes' o 'no')
15. **higher** - Desea cursar educación superior - ('yes' o 'no')
16. **internet** - Acceso a Internet en el hogar - ('yes' o 'no')
17. **romantic** - En una relación romántica - ('yes' o 'no')

##### **Ordinales:**
1. **famrel** - Calidad de las relaciones familiares -  (de 1 - muy mala a 5 - excelente)
2. **freetime** - Tiempo libre después de la escuela - (de 1 - muy bajo a 5 - muy alto)
3. **goout** - Frecuencia de salida con amigos - (de 1 - muy bajo a 5 - muy alto)
4. **Dalc** - Consumo de alcohol entre semana - (de 1 - muy bajo a 5 - muy alto)
5. **Walc** - Consumo de alcohol los fines de semana - (de 1 - muy bajo a 5 - muy alto)
6. **health** - Estado de salud actual - (de 1 - muy bajo a 5 - muy alto)
7. **traveltime** - Tiempo de viaje de casa a la escuela - (1: <15 min, 2: 15 a 30 min, 3:30 min a 1 h, 4: >1 h)
8. **studytime** - Tiempo semanal de estudio - ( 1: <2 h, 2: 2 a 5 h, 3: 5 a 10 h, 4: >10 h)
9. **Medu** - Nivel de educación de la madre - (0 - sin educación, 1 - primaria, 2 - secundaria baja, 3 - secundaria alta, 4 - educación superior)
10. **Fedu** - Nivel de educación del padre - (0 - sin educación, 1 - primaria, 2 - secundaria baja, 3 - secundaria alta, 4 - educación superior)

### **Cuantitativas o Numéricas**

##### **Discretas:**
1. **age** - Edad del estudiante (de 15 a 22 años)
2. **failures** - Número de fallos en materias anteriores - entre 0 y 3
3. **absences** - Número de ausencias escolares - de 0 a 93 
4. **G1** - Nota del primer período - entre 0 y 20 
5. **G2** - Nota del segundo período - entre 0 y 20 
6. **G3** - Nota final - entre 0 y 20 

##### **Continuas:**
Ninguna. Todas las variables numéricas presentadas son discretas en este caso porque corresponden a conteos o valores categorizados en escalas específicas.

```python
def classify_grade(grade):
    if grade >= 16:
        return 'Excelente/Muy bueno'
    elif grade >= 14:
        return 'Bueno'
    elif grade >= 12:
        return 'Satisfactorio'
    elif grade >= 10:
        return 'Suficiente'
    else:
        return 'Suspendido'

df_mat['grade_class'] = df_mat['G3'].apply(classify_grade)
df_por['grade_class'] = df_por['G3'].apply(classify_grade)

# Verificar las primeras filas con las nuevas columnas
print(df_mat[['G3', 'grade_class']].head())
print(df_por[['G3', 'grade_class']].head())
```

2. **Análisis Descriptivo y Métricas (df_por)**

Con el fin de evitar repetir esta parte del analisis para ambos dataframes, se decidió realizarlo unicamente para **df_por** ya que es el que mayor cantidad de registros tiene.

2.1 Descripción de **variables cualitativas**:

```python
variables_cualitativas = [
    'school', 'sex', 'address', 'famsize', 'Pstatus', 
    'Mjob', 'Fjob', 'reason', 'guardian', 'schoolsup', 
    'famsup', 'paid', 'activities', 'nursery', 'higher', 
    'internet', 'romantic', 'famrel', 'freetime', 'goout', 
    'Dalc', 'Walc', 'health', 'traveltime', 'studytime', 
    'Medu', 'Fedu'
]

print(f"Variables cualitativas: {variables_cualitativas}")

# Tablas de frecuencias y gráficos para cada variable cualitativa
for col in variables_cualitativas:
    print(f"\nTablas de frecuencias para {col}:")
    
    # Calcular frecuencias
    frecuencias = df_por[col].value_counts()
    frecuencias_relativas = df_por[col].value_counts(normalize=True)
    acumuladas = frecuencias.cumsum()
    acumuladas_relativas = frecuencias_relativas.cumsum()
    
    tabla_frecuencias = pd.DataFrame({
        'Frecuencia Absoluta': frecuencias,
        'Frecuencia Relativa (%)': (frecuencias_relativas * 100).round(2),
        'Frecuencia Acumulada': acumuladas,
        'Frecuencia Acumulada (%)': (acumuladas_relativas * 100).round(2)
    })
    
    print(tabla_frecuencias)

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # Gráfico de barras
    axes[0].bar(frecuencias.index, frecuencias.values, color='skyblue', edgecolor='black')
    axes[0].set_title(f"Distribución de {col} - Barras")
    axes[0].set_xlabel(col)
    axes[0].set_ylabel("Frecuencia")
    axes[0].tick_params(axis='x', rotation=45)
    axes[0].grid(axis='y', linestyle='--', alpha=0.7)
    
    # Gráfico de torta
    axes[1].pie(
        frecuencias.values, labels=frecuencias.index, autopct='%1.1f%%',
        startangle=90, colors=plt.cm.Paired.colors
    )
    axes[1].set_title(f"Distribución de {col} - Torta")
    axes[1].axis('equal')  
    
    plt.tight_layout()
    plt.show()
```

2.2 Descripción de **variables cuantitativas (discretas)**:

```python
cuantitativas = ['age', 'failures', 'absences', 'G3']
print(f"Variables cuantitativas: {list(cuantitativas)}")

# Tablas de frecuencias y gráficos para cada variable cuantitativa
for col in cuantitativas:
    print(f"\nTablas de frecuencias para {col}:")
    # Calcular frecuencias
    frecuencias = df_por[col].value_counts()
    frecuencias_relativas = df_por[col].value_counts(normalize=True)
    acumuladas = frecuencias.cumsum()
    acumuladas_relativas = frecuencias_relativas.cumsum()
    
    tabla_frecuencias = pd.DataFrame({
        'Frecuencia Absoluta': frecuencias,
        'Frecuencia Relativa (%)': (frecuencias_relativas * 100).round(2),
        'Frecuencia Acumulada': acumuladas,
        'Frecuencia Acumulada (%)': (acumuladas_relativas * 100).round(2)
    })
    
    print(tabla_frecuencias)
    ## .-------------
    #print(df_por[col].value_counts().sort_index())
    
    # Histograma
    plt.figure(figsize=(8, 4))
    plt.hist(df_por[col].dropna(), bins=15, color="lightcoral", edgecolor="black")
    plt.title(f"Distribución de {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.tight_layout()
    plt.show()

# ECDF (Empirical Cumulative Distribution Function)
from statsmodels.distributions.empirical_distribution import ECDF

for col in cuantitativas:
    data = df_por[col].dropna()
    ecdf = ECDF(data)
    plt.figure(figsize=(8, 4))
    plt.step(ecdf.x, ecdf.y, where="post", label=f"ECDF de {col}", color="green")
    plt.title(f"ECDF de {col}")
    plt.xlabel(col)
    plt.ylabel("Probabilidad acumulada")
    plt.grid(axis="both", linestyle="--", alpha=0.7)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

Para el caso de **df_por**:

```python
# Calcular y mostrar estadísticas para la variable G3
variable = 'G3'

# 1. Rango y rango intercuartil
rango = df_por[variable].max() - df_por[variable].min()
q1 = df_por[variable].quantile(0.25)
q3 = df_por[variable].quantile(0.75)
rango_intercuartil = q3 - q1

print(f"Rango de {variable}: {rango}")
print(f"Rango Intercuartil (Q3 - Q1) de {variable}: {rango_intercuartil}")

# 2. Varianza y Desvío Estándar
varianza = df_por[variable].var()
desvio = df_por[variable].std()

print(f"Varianza de {variable}: {varianza}")
print(f"Desvío estándar de {variable}: {desvio}")

# 3. MAD: Mediana de las desviaciones absolutas
median = df_por[variable].median()
mad = np.median(np.abs(df_por[variable] - median))
print(f"MAD (Mediana de las desviaciones absolutas) de {variable}: {mad}")

# 4. Coeficiente de variación
media = df_por[variable].mean()
coef_var = (desvio / media) * 100
print(f"Coeficiente de variación de {variable}: {coef_var:.2f}%")

# 5. Skewness (Asimetría)
skewness = df_por[variable].skew()
print(f"Asimetría (Skewness) de {variable}: {skewness}")

# 6. BoxPlot y diagrama de violín
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# BoxPlot
axes[0].boxplot(df_por[variable], vert=False, patch_artist=True, 
                boxprops=dict(facecolor='skyblue', color='black'))
axes[0].set_title(f"BoxPlot de {variable}")
axes[0].set_xlabel(variable)

# Diagrama de violín
axes[1].violinplot(df_por[variable], showmeans=True, showmedians=True)
axes[1].set_title(f"Diagrama de Violín de {variable}")
axes[1].set_xlabel(variable)

plt.show()

# 7. Cálculo de outliers
outliers = df_por[(df_por[variable] < (q1 - 1.5 * rango_intercuartil)) | 
                  (df_por[variable] > (q3 + 1.5 * rango_intercuartil))]
print(f"Cantidad de outliers detectados en {variable}: {outliers.shape[0]}")
print("Valores de los outliers:")
print(outliers[variable])
```

**Resumen df_por**:

Las calificaciones finales G3 parecerían presentar una distribución moderadamente concentrada, con la mayoría de los estudiantes agrupados en valores cercanos a la mediana y notas altas más comunes que bajas (asimetría negativa). 

La variabilidad parece moderada viendo el IQR de 4 y desvío estándar de 3,23. 

La mayoría de las calificaciones se desvían en promedio 2 puntos en términos absolutos de la mediana (MAD).

Estos resultados sugieren un rendimiento académico mayoritariamente homogéneo, con excepciones puntuales de bajas calificaciones.

### Identificación de correlaciones
Vamos a intentar determinar si existen correlaciones entre las variables.

Para ello tenemos que trabajar con variable *numéricas*. Debemos convertir las variables que faltan a numéricas.

Vamos a aplicar *label encoding* a todas las columnas no numéricas. Para ello usaremos LabelEncoder de sklearn. 

```python
from sklearn.preprocessing import LabelEncoder

# Crear un diccionario para almacenar la codificación de cada columna
codificacion_columnas_mat = {}
codificacion_columnas_por = {}

# Identifica las columnas no numéricas
columnas_no_numericas_mat = df_mat.select_dtypes(exclude=['number']).columns
columnas_no_numericas_por = df_por.select_dtypes(exclude=['number']).columns

# Aplica Label Encoding a cada columna no numérica
label_encoder = LabelEncoder()

for col in columnas_no_numericas_mat:
    df_mat[col] = label_encoder.fit_transform(df_mat[col])
    # Guardar la codificación en el diccionario
    codificacion_columnas_mat[col] = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

for col in columnas_no_numericas_por:
    df_por[col] = label_encoder.fit_transform(df_por[col])
    # Guardar la codificación en el diccionario
    codificacion_columnas_por[col] = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))


df_codificacion_mat = pd.DataFrame.from_dict(codificacion_columnas_mat, orient='index').transpose()
df_codificacion_por = pd.DataFrame.from_dict(codificacion_columnas_por, orient='index').transpose()


# Muestra el DataFrame para verificar que las columnas no numéricas se transformaron correctamente
print(df_mat.head())
print(df_por.head())

```
